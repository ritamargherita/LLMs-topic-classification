```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.lines import Line2D
```

```python
df = pd.read_csv('/Users/brain/Documents/GitHub/LLMs-topic-classification/results/statistics/summary_statistics_for_plots.csv')
df = df.drop(columns='dataset')
```

```python
melted_df = pd.melt(df, id_vars=['context', 'system', 'softness'], 
                    var_name='countTypes', value_name='count')
```

```python
# Add a dark grid
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

custom_order = ["specific_terms_count", "general_terms_count", "other_terms_count", "unassigned_count", "wrong_assigned_count"]

sns.boxplot(x="countTypes",
            y="count",
            hue="system",
            data=melted_df,
            order=custom_order,
            palette="colorblind",
            flierprops=dict(marker='o', markersize=4, markerfacecolor='black'),
            width=0.5)

custom_labels = ["Specific", "General", "'Other'", "Unassigned", "Wrongly Assigned"]
plt.xticks(ticks=plt.xticks()[0], labels=custom_labels)

plt.xlabel("")  # Set the name of the x-axis
plt.ylabel("Occurrence (%)")  # Set the name of the y-axis

plt.legend(loc="upper right", fontsize="large")

plt.savefig('boxplot1.png')

plt.show()
```
