```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
```

```python
# independent variables
independent_vars = ['setting', 'softness', 'system', 'dataset']
```

```python
# dependent variables
dependent_vars = ['unassigned_count', 'wrong_assigned_count', 'general_terms_count',
                  'specific_terms_count', 'other_terms_count']
```

```python
# design matrix
X = df[independent_vars]
```

```python
# Perform MANOVA
manova = MANOVA.from_formula('unassigned_count + wrong_assigned_count + general_terms_count + specific_terms_count + other_terms_count ~ dataset + softness + system + setting', data=df)
```

```python
print(manova.mv_test())
```
