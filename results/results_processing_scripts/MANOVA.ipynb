```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.formula.api import ols
from statsmodels.multivariate.manova import MANOVA
```

```python
df = pd.read_csv('/Users/brain/Documents/GitHub/LLMs-topic-classification/results/statistics/summary_statistics_formatted_for_MANOVA.csv')
```

```python
df
```

```python
# NO_CONTEXT
```

```python
df_no_context = df[df["setting"]=="no-context"]
```

```python
# Perform MANOVA
manova = MANOVA.from_formula('unassigned_count + wrong_assigned_count + general_terms_count + specific_terms_count + other_terms_count ~  system', data=df_no_context)
```

```python
print(manova.mv_test())
```

```python
variables_list = ["unassigned_count", "wrong_assigned_count", "general_terms_count", "specific_terms_count", "other_terms_count"]
```

```python
# Perform Tukey's HSD test
for variable in variables_list:
    tukey_results_system = pairwise_tukeyhsd(df_no_context[variable], df_no_context['system'])
    print(variable, tukey_results_system)
```
